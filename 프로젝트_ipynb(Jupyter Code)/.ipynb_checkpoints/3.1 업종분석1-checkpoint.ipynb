```python
import pandas as pd
```

```python
df=pd.read_csv('서울시 우리마을가게 상권분석서비스(상권-추정매출).csv', encoding='cp949')
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib
font_location='c:/Windows/fonts/malgun.ttf'
font_name=font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font',family=font_name)
```

```python
df1=df[(df['상권_코드_명']=='명동(명동거리)')]
```

```python
df2=df1[(df1['기준_분기_코드']==1)][['서비스_업종_코드_명','점포수','분기당_매출_금액','분기당_매출_건수','주중_매출_비율']]
df3=df2[['서비스_업종_코드_명','점포수']]
df4=df3.sort_values('점포수', ascending=False)
df4
```

```python
df22=df1[(df1['기준_분기_코드']==2)][['서비스_업종_코드_명','점포수','분기당_매출_금액','분기당_매출_건수','주중_매출_비율']]
df32=df22[['서비스_업종_코드_명','점포수']]
df42=df32.sort_values('점포수', ascending=False)
```

```python
df23=df1[(df1['기준_분기_코드']==3)][['서비스_업종_코드_명','점포수','분기당_매출_금액','분기당_매출_건수','주중_매출_비율']]
df33=df23[['서비스_업종_코드_명','점포수']]
df43=df33.sort_values('점포수', ascending=False)
```

```python
df24=df1[(df1['기준_분기_코드']==4)][['서비스_업종_코드_명','점포수','분기당_매출_금액','분기당_매출_건수','주중_매출_비율']]
df34=df24[['서비스_업종_코드_명','점포수']]
df44=df34.sort_values('점포수', ascending=False)
```

```python
plt.figure(figsize=(14,7))
plt.bar(df4['서비스_업종_코드_명'], df4['점포수'],width=0.4)
plt.bar(df42['서비스_업종_코드_명'], df42['점포수'],width=0.4)
plt.bar(df43['서비스_업종_코드_명'], df43['점포수'],width=0.4)
plt.bar(df44['서비스_업종_코드_명'], df44['점포수'],width=0.4)
plt.xlabel('서비스업종', size=20)
plt.ylabel('점포수', size=20)
plt.xticks(rotation=90)
plt.show()
```
